## Let's look at Fourier series using `interactive`

A Fourier series is a function representation (or approximation)
for a periodic function useful
in many applications of mathematics such as electrical engineering.
One variety of fourier series for a function represents the function as a sum of
trigonometric curves:
$$
f(x) = \frac{1}{2} a_0 +
  \sum_{n = 1}^{\infty} 
    a_n \cos (n x) +
  \sum_{n = 1}^{\infty} 
    b_n \sin (n x)
$$
Here is a simplified version of a fourier series using just a few $a$s:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def fourier3(x, a1, a2, a3):
    return a1 * np.cos(x) + a2 * np.cos(2*x) + a3 * np.cos(3*x)

fourier3(1,2,3,4)
```

```python
# What does a fourier approximation look like?
def plot_fourier3(a1, a2, a3):
    low = 0
    high = 4 * np.pi
    x = np.linspace(low, high, 100)
    y = fourier3(x, a1, a2, a3)
    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.grid(True)
    plt.show()
    
plot_fourier3(1,2,3)
```

```python
# We can use interact to interact with the approximation
from ipywidgets import interact

interact(plot_fourier3, a1=(-5.1,5), a2=(-5.1,5), a3=(-5.1,5))
```
